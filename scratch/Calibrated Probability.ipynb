```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
pd.options.display.float_format = '{:,.1f}'.format
```

```python
from sklearn.datasets import make_classification
n_samples = 10000
n_bins = 10

X, y = make_classification(n_samples=n_samples, n_classes =2,
                           #n_features=10, n_informative=2, n_redundant=8,
                           n_features=10, n_informative=10, n_redundant=0,
                           shuffle=False, random_state=1)

# dfX = pd.DataFrame(data=X)
# dfy = pd.DataFrame(data=y)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.5)
n_test = y_test.size
```

```python
from sklearn.linear_model import LogisticRegressionCV
from sklearn.metrics import brier_score_loss
print("Training set size:",X_train.shape)
print("Classes count:", np.bincount(y_train))
lr = LogisticRegressionCV().fit(X_train, y_train)
```

```python
# Quick inspection
# probs  = lr.predict_proba(X_test)[:10]
# labels = np.array([y_test[:10]]).T
# dfA = pd.DataFrame(data=probs, columns = ['P(0)','P(1)'])
# dfB = pd.DataFrame(data=labels, columns = ['label'])
# dfC = pd.concat([dfA,dfB], axis=1)
# print(dfC)
```

```python
from sklearn.calibration import calibration_curve
probs = lr.predict_proba(X_test)[:, 1]

fraction_of_positives, mean_predicted_value = calibration_curve(y_test, probs, n_bins=n_bins)
# print("fraction_of_positives", fraction_of_positives)
# print("mean_predicted_value", mean_predicted_value)

# Brier score
clf_score = brier_score_loss(y_test, probs, pos_label=1)
print("\tBrier score: %1.3f" % (clf_score))
```

```python
fig, ax = plt.subplots(1, figsize=(12, 6))
ax.get_xaxis().set_minor_locator(mpl.ticker.AutoMinorLocator())
ax.get_yaxis().set_minor_locator(mpl.ticker.AutoMinorLocator())
ax.grid(b=True, which='major', color='slategray', linewidth=1.0)
# ax.grid(b=True, which='minor', color='gray', linewidth=0.5)
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
# plt.xticks(np.arange(0, 1.1, 0.1))
# plt.yticks(np.arange(0, 1.1, 0.1))

# Lines
plt.plot(mean_predicted_value, fraction_of_positives, 's-', color ="crimson", markersize=12)
plt.plot([0, 1], [0, 1], '--', color='darkgreen') 

# Bars
hist, bin_edges = np.histogram(probs[y_test==1], bins=n_bins, range= (0,1))
posClass = plt.bar( (bin_edges[1:] + bin_edges[:-1]) * .5, hist/n_test,
# posClass = plt.bar( (bin_edges[1:] + bin_edges[:-1]) * .5, hist/hist.sum(), # Fraction of class
    width=(bin_edges[1] - bin_edges[0]), color="darkolivegreen", alpha=0.7, label='Positive class')
print("# positives = ", hist.sum())
hist_pos = hist
print(hist_pos)

# Bars
hist, bin_edges = np.histogram(probs[y_test==0], bins=n_bins, range= (0,1))
negClass = plt.bar( (bin_edges[1:] + bin_edges[:-1]) * .5, hist/n_test, # Fraction of total
#negClass = plt.bar( (bin_edges[1:] + bin_edges[:-1]) * .5, hist/hist.sum(), # Fraction of class
    width=(bin_edges[1] - bin_edges[0]), color="pink", alpha=0.7, label='Negative class')
print("# negatives = ", hist.sum())
hist_neg = hist
print(hist_neg)

# Verified fraction_of_positives = hit_rate
# hit_rate = hist_pos/(hist_pos+hist_neg)
# plt.plot(mean_predicted_value, hit_rate, 's-', color ="darkmagenta", markersize=12)

# plt.legend(handles=[posClass,negClass])

plt.xlabel("mean predicted value")
plt.ylabel("fraction of positives (hit rate)")
plt.title("$LogisticRegression$ Calibration Curve", fontsize=20); pass
```
